# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#display plots in notebook itself
%matplotlib inline
```

## Make a simple plot

```python
height = [150,160,165,185]
weight = [70,80,90,100]

#draw the plot
plt.plot(height,weight)

#add title
plt.title("Relationship between height and weight")

#label axis
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
#draw the plot for calories burnt
calories_burnt = [65,75,95,99]

#figure size
plt.figure(figsize=(15,5))

#Setting Colors
plt.plot(calories_burnt,'go')

#draw the plot for weight
plt.plot(weight, 'y--')

#Adding legends in lower-right part of the figuare
plt.legend(labels=['Calories Burnt', 'Weight'], loc='lower right')

#set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4'])

```

### Figuares and subplots


```python
#create 2 plots
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))

#plot on 0 row and 0 column
ax[0,0].plot(calories_burnt, 'bo')

#plot on 0 row and 1 column
ax[0,1].plot(weight)

#set titles for subplots
ax[0,0].set_title("Calories_Burnt")
ax[0,1].set_title("Weight")

#set ticks for each of these persons
ax[0,0].set_xticks(ticks=[0,1,2,3])
ax[0,1].set_xticks(ticks=[0,1,2,3])

#set labels for each of these persons
ax[0,0].set_xticklabels(labels=['p1','p2','p3','p4'])
ax[0,1].set_xticklabels(labels=['p1','p2','p3','p4'])
```

```python
# when you want a common axis then use sharey=True and rows =1

#create 2 plots
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(6,6), sharey=True)

#plot on 0 row and 0 column
ax[0].plot(calories_burnt, 'go')

#plot on 0 row and 1 column
ax[1].plot(weight)

#set titles for subplots
ax[0].set_title("Calories_Burnt")
ax[1].set_title("Weight")

#set ticks for each of these persons
ax[0].set_xticks(ticks=[0,1,2,3])
ax[1].set_xticks(ticks=[0,1,2,3])

#set labels for each of these persons
ax[0].set_xticklabels(labels=['p1','p2','p3','p4'])
ax[1].set_xticklabels(labels=['p1','p2','p3','p4'])
```

# Data Visualization using matplotlib

```python
#read the dataset
data_BM = pd.read_csv('bigmart_data.csv')

#drop the null values
data_BM = data_BM.dropna(how="any")

#view the top results
data_BM.head()
```

## Line Chart

```python
price_per_item = data_BM.groupby('Item_Type').Item_MRP.mean()[:10]
price_per_item
```

```python
x = price_per_item.index.tolist()
y = price_per_item.values.tolist()

#set figuare size
plt.figure(figsize=(14,8))

#set title
plt.title('Mean price for each item type')

#set axis labels
plt.xlabel('Item_type')
plt.ylabel('Mean Price')

#set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.plot(x,y)
```

## Bar Chart

```python
#sales by outlet size
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

#sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

#set axis labels
plt.xlabel('Outlet Size')
plt.ylabel('Sales')

#set title
plt.title('Mean sales for each outlet size')

#set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x,y, color=['red', 'orange', 'magenta'])
```

## Histograms

```python
#set title
plt.title('Item MRP(price) distribution')

#set label
plt.xlabel('Item_MRP')
plt.ylabel('Frequency')

#plot histogram
plt.hist(data_BM['Item_MRP'], bins=20, color='blue')
```

## Box Plots

```python
data = data_BM[['Item_Outlet_Sales']]

#create outlier point shape
red_diamond = dict(markerfacecolor = 'r', marker='D')

#set title
plt.title('Item Sales Distribution')

#make the box plot
plt.boxplot(data.values, labels=['Item Sales'], flierprops=red_diamond)
```

```python
# Creating multiple box plot 

data1 = data_BM[['Item_Weight','Item_MRP']]
#generate subplots
fig,ax = plt.subplots()

#create outlier point shape
red_diamond = dict(markerfacecolor = 'r', marker='D')

#set title
plt.title('Item Sales Distribution')

#make the boxplot
plt.boxplot(data1.values, labels=['Item Weight', 'Item MRP(price)'])
```

## Violin Plots

```python
#add labels to x axis
plt.xticks(ticks=[1,2], labels=['Item Weight', 'Item MRP'])

#make the violin plot
plt.violinplot(data1.values)
```

## Scatter Plots

```python
## Scatter Plots
plt.xlabel('Item Weight')
plt.ylabel('Item_Visibility')

plt.scatter(data_BM["Item_Weight"][:200],
           data_BM["Item_Visibility"][:200])
```

## Bubble Plots

```python
plt.xlabel('Item_MRP')
plt.ylabel('Item_Outlet_Sales')

plt.title('Item Outlet Sales vs Item MRP(price)')

plt.scatter(data_BM["Item_MRP"][:100],data_BM["Item_Weight"][:100], s=data_BM["Item_Visibility"][:100] *1000)
```

# Basics Of Seaborn

```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set(style = "darkgrid")

import warnings
warnings.filterwarnings("ignore")
```

```python
#read the dataset
data_BM = pd.read_csv('bigmart_data.csv')

#drop the null value
data_BM = data_BM.dropna(how="any")

#multiply Item Visibility by 100 to increase size
data_BM["Visibility_Scaled"] = data_BM["Item_Visibility"]*100

#view the top results
data_BM.head()
```

## Creating Basic Plots

### Line Plots

```python
sns.lineplot(x="Item_Weight", y="Item_MRP", data=data_BM[:50])
```

### Barchart

```python
sns.barplot(x="Item_Type", y="Item_MRP", data=data_BM[:5])
```

### Histogram

```python
sns.distplot(data_BM["Item_MRP"])
```

### Box Plot

```python
sns.boxplot(data_BM['Item_Outlet_Sales'], orient='vertical')
```

### Violin Plot

```python
sns.violinplot(data_BM['Item_Outlet_Sales'], orient='vertical',color='magenta')
```

### Scatter Plot

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=data_BM[:200], kind="scatter")
```

### Hue Semantic

```python
# Adding a third dimension

sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type", data=data_BM[:200])
```

## Different lineplots for different categories of the outlet size

### Line Plots

```python
sns.lineplot(x="Item_Weight", y="Item_MRP", hue="Outlet_Size", data=data_BM[:150])
```

### Bubble Plots

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales",data=data_BM[:200], kind="scatter",size="Visibility_Scaled", hue="Visibility_Scaled")
```

```python
# Plots for each category of Outlet Size
sns.relplot(x="Item_MRP", y="Item_Visibility",hue="Outlet_Size", style="Outlet_Size", col="Outlet_Size", data=data_BM[:100])
```
